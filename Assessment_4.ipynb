**Removing irrelevant features**

```python
df=df.drop(['id'], axis=1) 
```

```python
# Map "M" and "B" to 1 and 0 

df['diagnosis']= df['diagnosis']  .map({'M': 1, 'B': 0})
df
```

```python
# Outliers

threshold = 3  
outliers = pd.DataFrame()

df_copy = df.copy()

for col in df_copy.columns:
    # z-score
    df_copy[f'z_score_{col}'] = (df_copy[col] - df_copy[col].mean()) / df_copy[col].std()
    
    # mark outliers
    df_copy[f'Outlier_{col}'] = np.where(np.abs(df_copy[f'z_score_{col}']) > threshold, True, False)

    # record outliers
    outliers = pd.concat([outliers, df_copy[df_copy[f'Outlier_{col}'] == True][[col, f'z_score_{col}']]], axis=0)


outliers_index = df_copy.loc[df_copy[[f'Outlier_{col}' for col in df.columns]].any(axis=1)].index
outliers_index
```

```python
df_no_outliers = df.drop(outliers_index)
df_no_outliers
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Function to evaluate a model
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred, output_dict=True)
    return {
        "accuracy": accuracy_score(y_test, y_pred),
        "precision": report['weighted avg']['precision'],
        "recall": report['weighted avg']['recall'],
        "f1-score": report['weighted avg']['f1-score']
    }

# Load the dataset
data = pd.read_csv('C://Users//anon//Desktop//Duplicate//breast-cancer-wisconsin-data_data.csv')

# Preprocessing
data = data.drop(columns=['id'], errors='ignore')  # Drop unnecessary columns
X = data.drop(columns=['diagnosis'])
y = data['diagnosis'].map({'B': 0, 'M': 1})  # Encode target variable

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define models
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(),
    "Artificial Neural Network (ANN)": MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, activation='relu', solver='adam'),
}

# Store results
results = {}

# Evaluate each model
for name, model in models.items():
    results[name] = evaluate_model(model, X_train, X_test, y_train, y_test)

# Evaluate LSTM separately
X_train_lstm = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test_lstm = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

lstm_model = Sequential()
lstm_model.add(LSTM(100, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
lstm_model.add(Dropout(0.2))
lstm_model.add(Dense(1, activation='sigmoid'))
lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the LSTM model
lstm_model.fit(X_train_lstm, y_train, epochs=10, batch_size=32, verbose=0)

# Predict with LSTM
y_pred_lstm = (lstm_model.predict(X_test_lstm) > 0.5).astype(int)
lstm_report = classification_report(y_test, y_pred_lstm, output_dict=True)

# Add LSTM results to the results dictionary
results['LSTM'] = {
    "accuracy": accuracy_score(y_test, y_pred_lstm),
    "precision": lstm_report['weighted avg']['precision'],
    "recall": lstm_report['weighted avg']['recall'],
    "f1-score": lstm_report['weighted avg']['f1-score']
}

# Convert results to DataFrame and display
results_df = pd.DataFrame(results).T
print(results_df)
```
